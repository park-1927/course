# [파이썬으로 주성분 분석 구현](https://)
- PCA는 최소 차원을 사용하여 데이터의 분산을 최대한 보존하는 선형 변환을 사용

```python
import pandas as pd
import numpy as np

# Here we are using inbuilt dataset of scikit learn
from sklearn.datasets import load_breast_cancer

# instantiating
cancer = load_breast_cancer(as_frame=True)
# creating dataframe
df = cancer.frame

# checking shape
print('Original Dataframe shape :',df.shape)

# Input features
X = df[cancer['feature_names']]
print('Inputs Dataframe shape   :', X.shape)
df.head()
```

- 데이터 표준화를 적용 - 이를 위해 먼저 피처 공간에서 각 피처의 평균과 표준 편차를 계산

```python
# Mean
X_mean = X.mean()

# Standard deviation
X_std = X.std()

# Standardization
Z = (X - X_mean) / X_std

print(X_mean)
print(X_std)
print(Z)
```

- 공분산 행렬을 통해 두 특성의 종속성이 특성 공간에서 얼마나 강한지를 시각화 활용

```python
# covariance
c = Z.cov()

# Plot the covariance matrix
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(c)
plt.show()
```

- 특징 공간에 대한 고유 벡터와 고유 값을 계산 - 이는 특징 공간의 주성분을 식별하는 데 큰 역할을 함

```python
eigenvalues, eigenvectors = np.linalg.eig(c)
print('Eigen values:\n', eigenvalues)
print('Eigen values Shape:', eigenvalues.shape)
print('Eigen Vector Shape:', eigenvectors.shape)
```

- 고유값을 내림차순으로 정렬하고, 그에 따라 고유벡터도 정렬

```python
# Index the eigenvalues in descending order
idx = eigenvalues.argsort()[::-1]

# Sort the eigenvalues in descending order
eigenvalues = eigenvalues[idx]

# sort the corresponding eigenvectors accordingly
eigenvectors = eigenvectors[:,idx]
```

- 설명된 분산은 원래의 특징 공간 대신 ​​주성분을 선택함으로써 유지된 총 분산의 양에 대한 아이디어를 제공하는 용어

```python
explained_var = np.cumsum(eigenvalues) / np.sum(eigenvalues)
explained_var
```

- 주성분의 수를 결정하세요
>- 여기서는 원하는 값의 주성분 수를 고려하거나, 설명된 분산을 극한으로 설정할 수 있음
>- 여기서는 설명된 분산이 50% 이상인 경우를 가정
>- 여기에 몇 개의 주성분이 포함되는지 확인!!!

```python
n_components = np.argmax(explained_var >= 0.50) + 1
n_components
```

- 선택된 주성분에 데이터 투영
>- PCA는 전체 (x, y) 좌표를 저장하는 대신 주성분을 따라 투영 값만 저장하여 데이터 처리를 간소화
>- 투영 행렬 : 데이터의 공분산 행렬의 가장 큰 고유값에 해당하는 고유 벡터로 구성된 행렬
>- 고차원 데이터 세트를 저차원 부분 공간에 투영

```python
# PCA component or unit matrix
u = eigenvectors[:,:n_components]
pca_component = pd.DataFrame(u,
                             index = cancer['feature_names'],
                             columns = ['PC1','PC2']
                            )

# plotting heatmap
plt.figure(figsize =(5, 7))
sns.heatmap(pca_component)
plt.title('PCA Component')
plt.show()
```

- 다음 다음 공식을 사용하여 데이터 세트를 투영

image.png

- 주성분 u (녹색 벡터)는 데이터 분산을 최대화하고 투영의 새로운 축 역할을함
- 데이터 점 P1(x1,y1) (빨간색 벡터)은 원래 관측값이며, u (파란색 선)에 대한 투영은 축소된 차원에서 변환된 좌표를 나타냄
- 이 투영은 주요 특성을 유지하면서 데이터를 단순화함
>- 데이터의 공분산 행렬의 고유 벡터를 데이터의 주축이라고 하며, 이러한 주축에 대한 데이터 인스턴스의 투영을 주성분이라고 함

```python
# Matrix multiplication or dot Product
Z_pca = Z @ pca_component
# Rename the columns name
Z_pca.rename({'PC1': 'PCA1', 'PC2': 'PCA2'}, axis=1, inplace=True)
# Print the  Pricipal Component values
print(Z_pca)
```

- 대부분의 분산을 설명하는 축(차원)만 유지하고 다른 모든 축을 버림으로써 차원 축소를 얻음

# [Sklearn을 사용한 PCA](https://)
- 주성분 분석의 전체 과정을 자동화하는 여러 라이브러리가 있음
- 패키지에 함수 형태로 구현하여 원하는 주성분의 개수만 전달하면 됨
- Sklearn은 아래와 같이 PCA에 사용할 수 있는 라이브러리 중 하나임

```python
# Importing PCA
from sklearn.decomposition import PCA

# Let's say, components = 2
pca = PCA(n_components=2)
pca.fit(Z)
x_pca = pca.transform(Z)

# Create the dataframe
df_pca1 = pd.DataFrame(x_pca,
                       columns=['PC{}'.
                       format(i+1)
                        for i in range(n_components)])
print(df_pca1)
```

- 위의 Z_pca 결과는 정확히 동일한 값을 가짐

```python
# giving a larger plot
plt.figure(figsize=(8, 6))

plt.scatter(x_pca[:, 0], x_pca[:, 1],
            c=cancer['target'],
            cmap='plasma')

# labeling x and y axes
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```
