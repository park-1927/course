#[가중치(weight)](https://)

- [각 입력 신호가 결과 출력에 미치는 중요도를 조절하는 매개변수(parameter)](https:// )
>- 입력 값에 곱해지는 수이며, 입력으로 들어온 데이터들 중에서 [어떤 feature를 많이 반영하고, 어떤 feature를 덜 반영할지 결정해줌](https://)
>>- 데이터를 각기 다른 비중으로 다음 은닉층(hidden layers)으로 전달시키기 위해 가중치(weight), 즉, 중요도를 다르게 함
>>- 처음 들어오는 데이터(입력층)에서 다음 노드로 넘어갈때 모두 같은 값이면  
계속 같은 값이 나옴
>>- 각기 다르게 곱해야 함
>- [가중치는 활성화 함수에 따라 기울기를 증가시킴](https://)
>- [가중치가 커질 수록 그 feature는 모델에 더 많은 영향을 미치게 됨](https://)
- 예)
>- 집값을 예측하는 모델이 있다고 가정
>>- 입력으로 들어오는 feature는 역과의 거리, 방 넓이, 층 수, 창문의 개수 등이 있음
>>- 이 중 창문의 개수는 다른 세 개의 feature에 비해 덜 중요함
>>- 따라서 [역과의 거리, 방 넓이와 층 수는 높은 가중치를 창문의 개수는 낮은 가중치를 가지도록 학습됨](https://)

#[편향(Bias)](https://)
- [뉴런의 활성화 조건을 결정하는 매개변수](https://)
>- 활성화 함수에서 설정된 임계값을 얼마나 쉽게 넘게 할 것인지 결정해줌
>- 편향이 높다면 그만큼 활성화 함수의 임계값을 넘기 어렵기 때문에 까다로운 모델이 됨
>>- 활성화 함수의 임계값을 넘지 못하면 그 뉴런은 활성화되지 못하기 때문에 간단한 모델이 됨
>>- 편향이 작으면 활성화되는 뉴런이 많아져 복잡한 모델이 됨

- [활성화 함수는 0을 기준으로 x가 음수 값일때 0 또는 음수값의 출력을 내고   
x가 양수값일때 양수 값의 출력을 냄](https://)
- y=wx+b라는 함수가 있다고 했을 때 wx+b가 양수가 되어 뉴런에 활성화를 일으키고 싶다면 b가 wx보다 커야 할 것임
- 편향이 높다면 그만큼 기준이 까다롭기 때문에(전체적인 임계값이 높아지기에) 활성화되는 뉴런이 적어져 모델이 간단해짐
- 활성화를 시키고싶지 않은 경우에도 동일함
- [아래 그림에서 확인할 수 있듯이 모델의 결정 경계의 위치를 수평적으로 변경함](https://)
>- 하나의 뉴런으로 입력된 모든 값을 다 더한 다음에(가중합) 이 값에 더 해주는 상수
>- [하나의 뉴런에서 활성화 함수를 거쳐 최종적으로 출력되는 값을 조절하는 역할을 함](https://)
>>- 파티에 가는것을 무척 좋아하는 '성향'을 가진 A와 그렇지 않은 B라는 두사람이 있다면 날씨, 친구, 비용의 중요도(=가중치)같다고 가정하자!!
>>- A는 파티를 너무 좋아해서 돈이 없거나 날씨가 안좋아도 무조건 가려고 할테고 반면 B는 날씨도 좋고 돈도있는데도 조용한것을 더 좋아해서 파티에 안가려고 할수도 있는 경우
>>- 똑같이 값을 내면 안되겠죠? bias이라는 값을 추가 해주는 것!!!!

- 위 활성화 함수에 통과시키면 각각 0.0488와  0.9512이라는 값이 나옴
>- [편향의 값이 -1.3에서 4로 바뀌었을 뿐인데 0에 가깝던 값이 1에 가까운 값으로 변함](https://)
>- [같은 가중치를 가져도 편향 값에 따라 최종 출력은 다양해지기 때문에 뉴런의 활성화 조건을 결정해줄 수 있음](https://)

```python
import numpy as np
def logistic_function(x):
    return 1 / (1+np.exp(-x))

x1 = logistic_function(-2.97)
x2 = logistic_function(2.33)
print(x1)
print(x2)
```
